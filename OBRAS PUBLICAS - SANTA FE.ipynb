# Gráfico de caja para comparar los montos gastados en cada departamento

----
# gráfico de barras que muestra la distribución de montos gastados por cada responsable

```python
plt.figure(figsize=(12,6))
sns.barplot(x="responsable" , y="Monto Gastado" , data=data , ci=None)
plt.xticks(rotation=90 , ha="right")
plt.title("Distribución de Montos Gastados por responsable de obra")
plt.show()
```

# agrupamos los datos por responsable y muestra estadísticas descriptivas, para los montos gastados por cada responsable.

```python
# Estadísticas descriptivas por responsable
estadisticas_por_responsable = data.groupby('responsable')['Monto Gastado'].describe()

# Visualizar estadísticas descriptivas
print(estadisticas_por_responsable)
```

```python
# Identificación de obras con montos gastados atípicos
outliers = data[data['Monto Gastado'] > data['Monto Gastado'].quantile(0.75) + 1.5 * (data['Monto Gastado'].quantile(0.75) - data['Monto Gastado'].quantile(0.25))]

# Mostrar obras con montos gastados atípicos
print("Obras con Montos Gastados Atípicos:")
print(outliers[['Departamento', 'tipo de obra', 'responsable', 'Monto Gastado']])
```

* ## En este ejemplo, se utilizó el método de Tukey para identificar outliers. Las obras con montos gastados que superan cierto umbral (calculado en base a los cuartiles) son consideradas atípicas.
----

```python
# Estadísticas descriptivas de obras con montos gastados atípicos
estadisticas_outliers = outliers.describe()

# Visualizar estadísticas descriptivas
print("\nEstadísticas de Obras con Montos Gastados Atípicos:")
print(estadisticas_outliers)
```

* ## estadísticas descriptivas específicas para las obras identificadas como outliers.
----

```python
limites = [0, 1000000, 5000000, 10000000, float('inf')]
etiquetas = ["Bajo" , "Medio" , "Medio Alto" , "Alto"]

#Crear una nueva columna 
data["Segmento del monto"] = pd.cut(data["Monto Gastado"], bins=limites , labels=etiquetas , right=False)
```

```python
# Analizaremos las caracteristicas de cada segmento.

#Grafico de barras
plt.figure(figsize=(12,6))
sns.countplot(x="Segmento del monto", hue="Departamento" , data=data)
plt.title("Distribución de Departamento en Cada Segmento de Monto Gastado")
plt.show()

# Gráfico de barras para la distribución de responsables en cada segmento
plt.figure(figsize=(12, 6))
sns.countplot(x='Segmento del monto', hue='responsable', data=data)
plt.title('Distribución de Responsables en Cada Segmento de Monto Gastado')
plt.show()
```
